```
from keras.models import Sequential
from keras.layers import *
```

```
# Lets build a Neural Network with Single Neuron
model = Sequential()
model.add(Dense(6,input_shape=(2,),activation='tanh'))
model.add(Dense(4,activation='tanh'))
model.add(Dense(2,activation='softmax')) #Output Neuron

model.compile(optimizer='adam',loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```
#Train, starts your gradient descent optimization (using adam optimizer)
from keras.utils import to_categorical

Y_oht = to_categorical(Y)
print(Y_oht.shape)
```

```
model.fit(X,Y_oht,epochs=500)
```

```
loss,acc = model.evaluate(X,Y_oht) #Returns Final Loss and Accuracy
print("Accuracy ",acc)
```

```
plot_decision_boundary(lambda x:np.argmax(model.predict(x),axis=1),X,Y)
```
